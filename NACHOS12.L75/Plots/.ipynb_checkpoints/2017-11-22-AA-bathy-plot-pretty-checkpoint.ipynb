```python
# import modules
import xarray as xr
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy.ma as ma
import numpy as np
import netCDF4 as nc4
from datetime import datetime
from scipy.io import netcdf
from netCDF4 import Dataset

%matplotlib inline
```

```python
file = '/media/extra/DATA/NACHOS12.L75/NACHOS12.L75-I/NACHOS12.L75_bathymetry_20S_80N_nfrontiercleaned_repeat3glorys_retouchbalt.nc'
f= Dataset(file, "r", format="NETCDF4")

bathy=f.variables['Bathymetry']
navlon=f.variables['nav_lon']
navlat=f.variables['nav_lat']
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature

fig = plt.figure(figsize=(30,20))

ax = fig.add_subplot(111,projection=ccrs.PlateCarree(central_longitude=-30))
ax.autoscale(tight=True)
pcolor=ax.pcolormesh(navlon,navlat,bathy,transform=ccrs.PlateCarree(),cmap='gist_earth_r',vmin=0,vmax=6000)

ax.add_feature(cfeature.LAND,facecolor='grey')
ax.coastlines()
ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
ax.tick_params(labelsize=30)
fig.subplots_adjust(right=0.8)
ax.text(-0.07, 0.55, 'Latitude (in degree)', va='bottom', ha='center',
        rotation='vertical', rotation_mode='anchor',
        transform=ax.transAxes,fontsize=20)
ax.text(0.5, -0.2, 'Longitude (in degree)', va='bottom', ha='center',
        rotation='horizontal', rotation_mode='anchor',
        transform=ax.transAxes,fontsize=20)
cbar = plt.colorbar(pcolor,orientation='horizontal',shrink=0.75,label='m')
cbar.ax.tick_params(labelsize=30)
fig.suptitle("NATL60 bathymetry", fontsize=25)



plt.savefig('/media/extra/DATA/NACHOS12.L75/PLOTS/bathy_NACHOS12.png')
```

```python
cmap = plt.cm.gist_earth_r
cmap.set_under(color='grey')

plt.figure(figsize=(15,11))
ax = plt.subplot(111)
ax.autoscale(tight=True)
pcolor = ax.pcolormesh(navlon,navlat,
    bathy,cmap=cmap,vmin=10,vmax=6000,alpha=1)
ax.tick_params(labelsize=20)
cbar = plt.colorbar(pcolor,orientation='horizontal',pad=0.1)
cbar.ax.tick_params(labelsize=20)
ax.set_xlabel('Longitude (in degree)',fontsize=20)
ax.set_ylabel('Latitude (in degree)',fontsize=20)
cbar.ax.tick_params(labelsize=20)
plt.title("NATL60 bathymetry (in m)",fontsize=25)

plt.savefig('/home/albert/Data/NATL60/PLOTS/bathy_NATL60.png')
```
