# Zooms of bathymetry near Hudson Bay and Baltic Sea




## Hudson Bay

```python
%matplotlib inline
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
import numpy as npy

```

```python
file="/media/extra/DATA/NACHOS12/NACHOS12.L75-I/NACHOS12.L75_bathymetry_20S_80N_nfrontiercleaned_repeat3_retouchbalt.nc"

ds=xr.open_dataset(file)

bathy=ds.Bathymetry

lat=ds.nav_lat
lon=ds.nav_lon

```

```python
fig = plt.figure(figsize=(20,15))

ax1 = fig.add_subplot(111,projection=ccrs.PlateCarree(central_longitude=-30))
ax1.set_extent([-90, -50, 50, 80])
cmap = plt.cm.jet
cmap.set_under(color='grey')
pcolor=ax1.pcolormesh(lon,lat,npy.nan_to_num(bathy),transform=ccrs.PlateCarree(),cmap=cmap,vmin=15,vmax=8000)
ax1.add_feature(cfeature.LAND,facecolor='grey')
ax1.coastlines()
ax1.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.03, 0.7])
cbar=fig.colorbar(pcolor, cax=cbar_ax)
cbar.ax.tick_params(labelsize=20)
cbar.set_label("depth in m",size=17)

fig.suptitle("Hudson Bay", fontsize=25)

plt.savefig("bathys_nachos12_hudson_bay.png")
```
