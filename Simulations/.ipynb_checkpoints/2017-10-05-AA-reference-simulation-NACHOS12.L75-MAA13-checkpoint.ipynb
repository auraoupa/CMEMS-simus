## Configuration files

### Coordinates

- NACHOS12.L75_coordinates_20S_80N_v2.0.nc

From JMM's eORCA12_coordinates.nc, ncks -F -d x,2271,3879 -d y,1807,3374.


### Bathymetry

- NACHOS12.L75_bathymetry_20S_80N_nfrontiercleaned_repeat3glorys_retouchbalt.nc

From RBB's eORCA12_bathymetry_v2.4.nc, ncks -F -d x,2271,3879 -d y,1807,3374. Then, with BMGTools, filling the lakes, Red Sea, Black Sea, Pacific and Indian corners, Hudson Bay, Baltic Sea, Northern Greenland Fjords (crossed by the frontier). With personnal tool, ORCA025 bathymetry at the frontier, repeated for 3 levels.

## Boundary conditions : BDY for northern and southern frontier

- new-GLORYS2V4-NACHOS12.L75_(S/N)bdy(SSH/U/V/T/S)_y????.nc

From GLORYS2V4 monthly files, with personnal tool, slice the northern and southern frontier, repeat each colum 3 times et cut appropriately.
